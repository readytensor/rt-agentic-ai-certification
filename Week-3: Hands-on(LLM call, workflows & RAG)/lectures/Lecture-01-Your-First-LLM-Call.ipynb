# Import necessary libraries

```python
from langchain_groq import ChatGroq # For using Groq LLM
from langchain_core.messages import HumanMessage, SystemMessage # For setting system prompt and human prompt
```

# Initialize the Groq chat model

```python
llm = ChatGroq(
    model="llama-3.1-8b-instant",  # You can also use "mixtral-8x7b-32768" or other models
    temperature=0.7,
    api_key="put your groq api key here"
)
```

## **Basic LLM Call**

```python
# Define our messages
messages = [
    SystemMessage(content="You are a helpful AI assistant."),
    HumanMessage(content="What is artificial intelligence?")
]

# Make the LLM call
response = llm.invoke(messages)

# Print the response
print(response)
```

As you can see, the output contains not just the actual response from the LLM, but also a lot of metadata — including the number of prompt and completion tokens used, the model name, response time, and other internal details. This information can be useful for monitoring performance, tracking costs, or debugging your application.

However, in most cases, users are primarily interested in the actual content generated by the model. So, let’s extract and print just the `content` field, which contains the message intended for the user.

```python
# Print the main response
print(response.content)
```

In this basic example above:
1. We initialize a `ChatGroq` instance with a specific model and temperature. Temperature determines how creative or deterministic the model will be. A lower temperature (close to 0) makes responses more focused, deterministic, and conservative. A higher temperature (close to 1) makes responses more creative, diverse, and sometimes unpredictable.
2. We create a list of messages to send to the model
3. We call `invoke()` with these messages
4. We print the response content

## **Working with Streaming Responses**

For longer responses, it's often better to stream the output as it's generated. Streaming means receiving and displaying the response token by token in real-time, rather than waiting for the complete response before showing anything.
If you've used AI chat services like ChatGPT, you've experienced streaming because when you ask a question, you don't get the entire answer at once. Instead, you see the text appearing word by word or phrase by phrase as if the AI is "thinking" and writing in real-time. This provides a much better user experience than waiting for a long response to complete before seeing anything.

```python
messages = [
    SystemMessage(content="You are a helpful AI assistant."),
    HumanMessage(content="Write a short story about the history of python programming language.")
]

# Stream the response
print("Streaming response:")
for chunk in llm.stream(messages):
    print(chunk.content, end="")
```

Notice how the content was being printed out? That's is exactly what streaming does. It doesn't show you the whole content once but it shows them word by word or sentence by sentence as they are being generated.

Streaming is particularly useful when:
- You want to show users immediate feedback
- You're generating longer content
- You want to process the response as it arrives

## **More Complex Interactions**

Let's build a slightly more complex example with a multi-turn conversation. Think about how you use ChatGPT or similar AI assistants - you might first ask "How do I create a list in Python?" and then follow up with "How can I sort it?" without needing to repeat that you're talking about Python lists. The AI remembers the context of your conversation.
To achieve this contextual memory in our own applications, we need to maintain the conversation history between LLM calls:

```python
from langchain_core.messages import AIMessage # For setting the AI message

# Initialize our conversation history
conversation = [
    SystemMessage(content="You are a knowledgeable AI assistant who specializes in Python programming."),
]

# User's first question
conversation.append(HumanMessage(content="How do I read a CSV file in Python?"))
response = llm.invoke(conversation)
print(f"User's first question: How do I read a CSV file in Python?\n")
print(f"AI's first response: {response.content}\n")

# Add the AI's response to the conversation history
conversation.append(AIMessage(content=response.content))

# User's second question - notice we don't need to mention CSV again
conversation.append(HumanMessage(content="How can I filter rows where a specific column has values greater than 100?"))
response = llm.invoke(conversation)
print(f"User's second question: How can I filter rows where a specific column has values greater than 100?\n")
print(f"AI's second response: {response.content}\n")
```

## **Using Structured Output**
One powerful feature of LangChain is the ability to get structured outputs from LLMs. This means instead of getting raw text responses that you have to parse yourself, you can define the exact data structure you want and receive the response as a properly formatted object.

# Why Structured Output Matters
Structured output is crucial for building reliable AI applications for several reasons:

1. Consistency: Ensures responses follow a predictable format
2. Validation: Automatically checks that all required fields are present and of the correct type
3. Integration: Makes it easy to pass LLM outputs directly into your application logic
4. Error Reduction: Minimizes parsing errors and edge cases in handling free-form text

Let's see how we can use this to extract specific information:


```python
from pydantic import BaseModel, Field # For defining the structure of the output
from typing import List # For defining the type of the output

class RecipeSummary(BaseModel):
    """Information about a cooking recipe."""
    title: str = Field(description="The name of the dish")
    ingredients: List[str] = Field(description="List of required ingredients")
    prep_time_minutes: int = Field(description="Preparation time in minutes")
    difficulty: str = Field(description="Difficulty level (Easy, Medium, Hard)")

# Create a structured model
structured_llm = llm.with_structured_output(RecipeSummary) 

# Ask for a recipe
recipe_query = "Give me a recipe for Nigerian Jollof Rice."
recipe_info = structured_llm.invoke(recipe_query)

print(f"Recipe: {recipe_info.title}")
print("Ingredients:")
for ingredient in recipe_info.ingredients:
    print(f"- {ingredient}")
print(f"Prep Time: {recipe_info.prep_time_minutes} minutes")
print(f"Difficulty: {recipe_info.difficulty}")
```

In this example, we've defined a Pydantic model that specifies exactly what information we want about a recipe (The popular Nigerian Jollof Rice). The LLM will structure its response to match this schema, making it easy to work with programmatically.

# Under the Hood
This functionality is powered by Pydantic, a data validation library that's become essential in the Python ecosystem. Pydantic models define both the structure of your data and validation rules, ensuring you get exactly what you expect.
The good news is that the Pydantic team has recognized the importance of this use case and has developed their own specialized framework called `pydantic-ai` that makes this process even more streamlined. While we're using LangChain's implementation today, we'll explore `pydantic-ai` in upcoming weeks when we start building more sophisticated agents.

# Practical Applications
Structured output is particularly valuable for:

- Extracting specific fields from documents (names, dates, amounts)
- Building database entries from unstructured content
- Creating searchable catalogs of information
- Converting free-form user requests into API-compatible parameters
- Ensuring consistency across multiple LLM calls in an agent workflow

By defining clear structures for your outputs, you make your AI applications more robust and easier to maintain.
