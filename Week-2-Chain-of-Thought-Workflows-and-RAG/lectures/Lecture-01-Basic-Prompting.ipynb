# **Basic Prompting**

## What is Prompting?

Prompting is the process of instructing an LLM on what to do by providing it with text-based instructions. It's essentially how we communicate with LLMs to get them to perform specific tasks or generate particular types of content. The quality and structure of your prompts significantly impact the quality of the responses you get from the model.

Think of prompting as giving instructions to a very intelligent but literal assistant who has vast knowledge but needs specific guidance on how to apply that knowledge to your particular needs.

## **Types of Prompts**

In most LLM applications, there are two main types of prompts that work together:

### **1. System Prompt**

The system prompt defines the overall behavior, role, and constraints for the LLM. It's typically invisible to the end user and sets up the "personality" and capabilities of the AI assistant. The system prompt is crucial because it:

- Establishes the context for the conversation
- Defines the specific role the LLM should take
- Sets boundaries on what the LLM should or shouldn't do
- Provides background information the LLM might need
- Defines the format or structure of expected responses

**Image you are building an AI powered Customer Support for Ready Tensor, the System Prompt for ReadyTensor Customer Support would be something like:**

```
You are ReadyTensor's AI customer support assistant. ReadyTensor is a pioneering media platform for artificial intelligence (AI) and data science projects, designed to accelerate innovation and showcase transformative technologies.

Key features of ReadyTensor include:
- Publications: A comprehensive space for AI developers, students, researchers, and practitioners to document, share, and collaborate on their work
- Jupyter-notebook-like interface for seamlessly organizing AI projects
- Customizable publication templates for comprehensive project documentation
- Collaboration tools including version control and issue tracking
- Flexible visibility options (public sharing to private, access-controlled publications)
- Scalable storage solutions for projects of any size
- Tools for reproducibility that make it easy to share detailed methodologies, code, and data

When responding to users:
1. Always be helpful, professional, and concise
2. If you're unsure about something, don't make up information - suggest contacting human support at support@readytensor.com
3. For technical questions about AI concepts, you can provide explanations, but focus on how these concepts apply to ReadyTensor's platform
4. When users report problems, ask for specific error messages or behaviors they're encountering
5. You can explain pricing plans (Free, Pro, and Team) but refer users to the website for the most current information
6. Don't share information about ReadyTensor's internal operations, codebase, or unreleased features

If users ask about competitors or comparisons, focus on ReadyTensor's strengths without disparaging competitors.

Format responses in a structured way with headings and bullet points when appropriate.
```

### **2. User Prompt**

The user prompt is the specific question, instruction, or request that the end user sends to the LLM. It's what the user types into the chat interface to interact with the AI assistant. The user prompt is typically much shorter than the system prompt and focuses on the immediate need or query.

**Example User Prompts for ReadyTensor Customer Support:**

```
How do I publish my AI project on ReadyTensor?
```

```
I'm getting an error when I try to upload my publication: "Storage limit exceeded". What does this mean?
```

```
Does ReadyTensor support Jupyter notebooks for data visualization?
```

```
What's the difference between your Free and Pro pricing plans?
```

## **Important Considerations When Defining System Prompts**

### **1. Token Limits**

Every LLM has a maximum number of tokens it can process in a single request. This limit includes both the system prompt, user prompt, and any context from previous messages. If you exceed this limit, the model might:

1. Truncate your input (losing important information) 
2. Reject the request entirely 
3. Charge you more for additional tokens (for commercial APIs)

**Examples of token limits for popular models:**
- GPT-3.5 Turbo: 4,096 tokens
- GPT-4: 8,192 tokens (some versions support 32k or more)
- Claude 3 Sonnet: 200,000 tokens
- Gemini Pro: 32,768 tokens

To manage token limits:
1. Keep system prompts concise and focused
2. Use bullet points rather than lengthy paragraphs
3. Remove unnecessary examples or redundant instructions
4. Consider implementing dynamic prompting that adapts based on the user's query
5. For complex applications, consider breaking the task into multiple API calls

### **2. Specificity and Clarity**

Be specific about what you want the LLM to do. Vague instructions lead to unpredictable results. Consider specifying:

1. The exact role the LLM should play
2. The format you want the response in
3. The tone or style of the response
4. Any assumptions the LLM should make
5. Information that should be included or excluded

### **3. Providing Examples (Few-shot Learning)**

Including examples of the desired input-output pairs can dramatically improve the quality of responses. This technique, called "few-shot learning," helps the LLM understand exactly what you're looking for.

**Example of Few-shot Learning in a System Prompt:**

```
When users ask about pricing, provide a structured response as shown below:

User: What are your pricing plans?
Assistant: 
# ReadyTensor Pricing Plans

## Free Plan - $0/month
- 10 published publications per month
- 10 publication assessments per month
- 1 GB storage for project files
- Perfect for students, hobbyists, and AI enthusiasts

## Pro Plan - $29.99/month
- 30 published publications per month
- 30 publication assessments per month
- 5 GB storage for project files
- Ideal for AI developers working on multiple projects

## Team Plan - $99.99/month
- Up to 5 members included
- 50 published publications per month
- 50 publication assessments per month
- 10 GB storage for project files
- 1 team hub (public/private publications)
- $19.99 per month for each additional member
- Designed for collaborative AI teams
```

### **4. Testing and Iteration**

System prompts should be thoroughly tested with various user queries and refined based on the responses. It's an iterative process to find the optimal prompt that produces the desired behavior consistently.
