# **Introduction to RAG (Retrieval Augmented Generation) systems**

## **Limitations of System Prompts: When LLMs Need More Knowledge**

While system prompts are powerful for defining an LLM's behavior and capabilities, they have a significant limitation: they cannot provide the LLM with information beyond what it learned during training. As we saw in our ReadyTensor example, a system prompt can effectively guide the AI to behave as a customer support agent, but it cannot help the AI answer questions about data it doesn't know.

For instance, if a user asks about ReadyTensor's newest feature released last week, or about specific customer account details, the LLM won't be able to provide accurate information unless this data was included in its training. Similarly, if company policies change after the LLM was trained, it would continue providing outdated information despite having a well-crafted system prompt.

This knowledge gap presents a real challenge when building practical AI applications. How can we enable LLMs to access and use the most current, relevant information without constantly retraining them? This is where Retrieval Augmented Generation (RAG) systems come into play.

# **RAG (Retrieval Augmented Generation)**

Since LLMs became popular and people started seeing things you can achieve with LLMs, there was one persistent issue: LLMs don't have access to the data needed to answer many real-world questions. They are limited to the information they were trained on, creating a significant knowledge cutoff problem.

LLMs are great but they can't answer questions that their training data doesn't cover. For instance, if you have an LLM that was trained in 2023 (meaning it has updated data until 2022), it can't answer current affairs questions in 2025 like "Who is the president of the USA?" The LLM would say Joe Biden because, as of the time of its training, Joe Biden was the president - not Donald Trump who is president now in 2025.

To solve this limitation, we need to find a way to give recent and relevant data to the LLM so it can answer questions using this up-to-date information rather than relying solely on its training data.

To achieve this, there are two major methods:
1. Finetuning of the LLM
2. RAG (Retrieval Augmented Generation)

**Finetuning** is the process of updating a pre-trained model with new data to make it learn new information or adapt to specific tasks. However, finetuning is not always sought for because:
1. It is costly
2. It is time and resource consuming
3. It needs so much technical know-how
4. If any data you used to finetune a model becomes outdated, you will need to repeat the entire finetuning process
5. It requires a substantial amount of data to be effective

Because of these and many more challenges faced with finetuning models, RAG came into place.

**RAG (Retrieval Augmented Generation) is a technique that combines information retrieval with text generation. It allows LLMs to access and use external knowledge that wasn't part of their original training data, enabling them to provide more accurate, up-to-date, and contextually relevant responses.**

Every RAG system is divided into two major parts:
- First part: Insertion Part
- Second part: Retrieval Part

### **First Part: The Insertion part**
This is the first stage of every RAG system. In this stage the document/documents which you want to use as your knowledge base will be chunked, embedded using any embedding method of your choice and uploaded to a vector database.

1. Chunking of documents: This is the process of dividing document into smaller chunks so that it can fit into the maximum input token acceptable by the embedding model. For example, let's say you have a single document with over 100,000 tokens, you can't pass this whole document through text-embedding-3-small which has a maximum input token of 8,191. You will need to chunk the document in a meaningful manner so that you can pass each chunk of the document to the embedding model. So in a nutshell, we chunk the documents so we can embed them and so that we can retrieve specific, relevant sections rather than entire documents.

2. Embedding the Chunked Document: After you must have chunked the document, the next step is to embed the document into vectors. Embedding is the process of converting text into numerical representations (vectors) that capture semantic meaning. It is a very necessary step because when data are stored as vectors, words that are closer in meaning to each other are positioned closer together in the vector space, making semantic search possible.

3. Uploading the vectors and document to a Vector Database: This is the last step in the first part of RAG system. After you must have embedded the document, you need to store both the chunks and vectors in a special type of database called vector database. Don't worry we will cover vector database and embedding models in the next section.

### **Second part: Retrieval Part**

After you have uploaded the necessary document which is your knowledge base, you can now query the database to get the answers you need.

Here is how it works:
When a user asks a question, the following steps will be taken:
1. The question will be embedded with the same embedding model you used to embed the initial document 
2. Then the embedded question will be sent to the vector database to make a similarity search on the vector database where the document is stored. (Similarity search is a type of search that finds items whose embeddings are closest to the query embedding, usually measured by cosine similarity or Euclidean distance)
3. After searching the database, the chunks of data related to the question will be retrieved from the database
4. This chunk of data retrieved will now be given to any LLM (Grok, GPT, Gemini etc) together with a system prompt which has been initially set
5. The LLM now answers the questions using the information given to it (the system prompt and the chunks)

### **Advantages of using RAG over finetuning**
1. You can see, unlike finetuning, you can always upload new documents into the vector database and you can easily remove outdated documents from the vector database
2. It's cheaper to run and maintain
3. It doesn't require so much technical know-how
4. It is not time consuming
5. It does not require high computing/resource power
6. It provides citation and source tracking, making responses more trustworthy and verifiable
7. It gives more control over the knowledge the model uses to generate responses

### **Popular use cases for RAG**
1. Customer support: Many companies have automated their customer service operations by building RAG systems that retrieve information from product manuals, FAQs, and previous customer interactions to provide accurate and helpful responses.
2. Legal research: Law firms use RAG to search through vast databases of legal documents, precedents, and case laws to assist lawyers in finding relevant information quickly.
3. Healthcare information systems: Medical professionals use RAG-based systems to access the latest research papers, treatment protocols, and patient history to aid in diagnosis and treatment planning.
4. Educational platforms: Online learning platforms implement RAG to provide students with personalized answers from textbooks, course materials, and educational resources.
5. Internal knowledge management: Organizations use RAG to help employees find information across company documents, policies, and procedures, making institutional knowledge more accessible.
6. Financial analysis: Financial institutions use RAG systems to analyze market reports, financial statements, and economic indicators to generate insights and forecasts.
7. Content recommendation: Media companies employ RAG to suggest relevant articles, videos, or products based on a user's interests and past interactions.
